```python
data = frame
co2_df = data.pivot(index='year', columns='country', values='co2')
gdp_df = data.pivot(index='year', columns='country', values='gdp')
co2_df = co2_df.fillna(0)#避免画饼图时出现错误，即分母为0
gdp_df = gdp_df.fillna(0)
co2_df
```

```python
co2_df.index = pd.to_datetime(co2_df.index,format='%Y')
gdp_df.index = pd.to_datetime(gdp_df.index,format='%Y')
bar = co2_df[tmp]#饼图，这里有支持的个数限制，我们这里取2020年前20的国家
```

```python
# 绘制逐年co2的排放，进行观察
co2_plot = co2_df.plot_animated(filename='每年co2排放.gif',title="CO2 emissions per year",n_visible=15 ,period_fmt="%Y",perpendicular_bar_func='mean')
```

观察我们的动态图，我们可以发现绝大部分时间中发达国家占统治地位，直到1950左右，以中国、印度和南非才在图表中出现，且在2000左右，整体的co2排放量才出现井喷式增长。\

#### 最后做一个有意思的图，通过链接进入

```python
tl = Timeline()
for i in range(1900, 2020,10):
    frame_time = frame[frame.year == i]
    map0 = (
        Map()
        .add("country",frame_time[['country','co2']].values.tolist(), "world")
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Map-{}年GDP（亿元）".format(i)),
            visualmap_opts=opts.VisualMapOpts(
                is_piecewise=True,
                pieces=[
                    {"min":0,"max":500,"label":"0~500","color":"cyan"},
                    {"min":500,"max":2000,"label":"500~2000","color":"yellow"},
                    {"min":2000,"max":5000,"label":"2000~5000","color":"orange"},
                    {"min":5000,"max":8000,"label":"5000~8000","color":"coral"},
                    {"min":8000,"max":120000,"label":"8000~12000","color":"red"},
                ]   ),))
    tl.add(map0, "{}年".format(i))
tl.render("1900~2019年全国各地区GDP.html")
```
